```python
# Barchart creation to understand the highest temperatures recorded by state
T1_maxtemp = T1.groupby('Country').AverageTemperature.max()
T1_maxtemp = T1_maxtemp[T1_maxtemp.values > 36]
xaxis = T1_maxtemp.index
value = T1_maxtemp.values
ax = T1_maxtemp.plot.bar(x=xaxis, y=value)
T1_maxtemp
```

```python
# Barchart creation to understand the lowest temperatures recorded by state
T1_mintemp = T1.groupby('Country').AverageTemperature.min()
T1_mintemp = T1_mintemp[T1_mintemp.values < -15]
xaxis = T1_mintemp.index
value = T1_mintemp.values
ax = T1_mintemp.plot.bar(x=xaxis, y=value)
T1_mintemp
```

```python
# Barchart creation to understand the highest temperatures recorded by City
T1_maxtempC = T1.groupby('City').AverageTemperature.max()
T1_maxtempC = T1_maxtempC[T1_maxtempC.values > 38]
xaxis = T1_maxtempC.index
value = T1_maxtempC.values
ax = T1_maxtempC.plot.bar(x=xaxis, y=value)
T1_maxtempC
```

```python
# Barchart creation to understand the lowest temperatures recorded by City
T1_mintempC = T1.groupby('City').AverageTemperature.min()
T1_mintempC = T1_mintempC[T1_mintempC.values < -33]
xaxis = T1_mintempC.index
value = T1_mintempC.values
ax = T1_mintempC.plot.bar(x=xaxis, y=value)
T1_mintempC
```

```python
Temperature = pd.read_csv('Temperature.csv')
Temperature
```

```python
Nazioni = Temperature.groupby('Country').Country.count()
Nazioni = Nazioni[Nazioni.values > 100000]
Nazioni
```

```python
# Piechart creation to understand the variation of workers according to study
xaxis = Nazioni.index
value = Nazioni.values
fig1, ax1 = plt.subplots(figsize=(15,15))
ax1.pie(value, labels=xaxis, autopct='%1.1f%%',  #'autopct' represent the field centre of labels
        shadow=False, startangle=90)
ax1.axis('equal')                                # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
